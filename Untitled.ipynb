```python
### Functions
def numberlist(nums,limit):   
    sum=0  
    for index,i in enumerate(nums):  
        sum += i
        if sum>=limit:  
            return nums[:index+1]
        
def plot_feature_importance(importance,names,model_type):
    
    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    
    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + 'FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
```

```python
### PCA ###
to_PCA = data[[ 'card_type', 'products_count', 'mnths_with_bank','mnths_inactive_last_year', 'contacts_last_year',\
             'education','martial_status', 'gender', 'earnings_bucket', 'age', 'family_size','cc_limit',\
             'trans_amount', 'trans_count', 'amount_change', 'count_change', 'revolving_balance', \
             'open_to_buy_cl', 'util_ratio']]
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(to_PCA)
# Apply transform to both the training set and the test set.
scaler_PCA = scaler.transform(to_PCA)
#test_df = scaler.transform(test_df)
from sklearn.decomposition import PCA
pca = PCA(n_components=19)
X_pca = pca.fit_transform(scaler_PCA)

features_to_PCA=[ 'card_type', 'products_count', 'mnths_with_bank','mnths_inactive_last_year', 'contacts_last_year',\
             'education','martial_status', 'gender', 'earnings_bucket', 'age', 'family_size','cc_limit',\
             'trans_amount', 'trans_count', 'amount_change', 'count_change', 'revolving_balance', \
             'open_to_buy_cl', 'util_ratio']
PC_values =  ['PC'+str(i) for i in range(1,len(features_to_PCA)+1)]
values = pca.explained_variance_ratio_
values = values.tolist()
dictionary = dict(zip(PC_values, values))
print(dictionary)
print(sum(values))
```

```python
variance_explained=0.95
### Feature importance 
importance = abs(pca.components_)
final_array = np.zeros(len(features_to_PCA), dtype =np.float64)
for i in range(len(features_to_PCA)):
    temp = np.zeros(len(features_to_PCA), dtype =np.float64)
    for j in range(len(features_to_PCA)):
        single_PC= importance[0]*values[j]
        temp = temp+single_PC
    final_array = final_array + temp
idx = (-final_array).argsort()[:len(features_to_PCA)]
```

```python
important_features = []
for i in range(len(idx)):
    feature = features_to_PCA[idx[i]]
    important_features.append(feature)
sorted_array = np.sort(final_array)
reverse_array = sorted_array[::-1]
result = reverse_array.tolist()
normalized_results = []
for r in range(len(result)):
    elem = (result[r]/sum(result))*100
    normalized_results.append(elem)
limit = sum(result)*variance_explained 
threshold = normalized_results[len(numberlist(result,limit))-1]
```

```python

### Print results ###
print("threshold={}".format(threshold))
print("normalized_results=", normalized_results)
print("important_features=", important_features)

fig=plt.figure(figsize=(15,10))
plt.title("Feature importance", fontsize=22 )
plt.xticks(rotation=90, fontsize=12)
plt.yticks( fontsize=12)
plt.xlabel("Features", fontsize=16)
plt.ylabel("Importance score", fontsize=16)

plt.bar(important_features, normalized_results,  width = 0.2, alpha=0.3, align='center', color='green')
plt.axhline(y=threshold, color='r', linestyle='--', label = 'blabla: {:.2f}'.format(threshold))
plt.grid(True)
plt.show()

result = dict(zip(important_features, normalized_results))
result = dict(sorted(result.items()))

d = dict((k, v) for k, v in result.items() if v > threshold)
print("List of features, which explain 95% of variance:", list(d.keys()))
```
